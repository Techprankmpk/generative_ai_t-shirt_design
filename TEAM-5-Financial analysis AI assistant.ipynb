```python
# Function to read text from PDF
def read_pdf(pdf_path):
    pdf_reader = PdfReader(pdf_path)
    raw_text = ''
    for page in pdf_reader.pages:
        content = page.extract_text()
        if content:
            raw_text += content
    return raw_text
```

```python
# Function to split text into chunks
def split_text(raw_text):
    chunk_size = 800
    chunk_overlap = 200
    chunks = [raw_text[i:i + chunk_size] for i in range(0, len(raw_text), chunk_size - chunk_overlap)]
    return chunks
```

```python
# Function to perform question answering
def perform_question_answering(docs, query):

    conversation_history = [{"role": "system", "content": "You are a helpful assistant. Make use of the following doc to answer the user. Doc:{}".format(docs)}]
    conversation_history.append({"role": "user", "content": query})
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-16k",
        temperature=0.6,
        presence_penalty=0,
        frequency_penalty=0,
        messages=conversation_history
    )

    return response["choices"][0]["message"]["content"]
```
